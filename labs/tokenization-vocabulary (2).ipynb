## Vocabulary

A mapping between a token and a unique ID number

Build one manually by keeping track of unique tokens

```python
tok2idx = {}
tok2idx["[UNK]"] = 0  # reserve one for unknown tokens during test time
next_idx = 1

for t in tokens:
    if t not in tok2idx:
        tok2idx[t] = next_idx
        next_idx += 1
```

```python
tok2idx
```

```python
VOC_SIZE = len(tok2idx)
VOC_SIZE
```

Then every token can now be converted into an int

```python
def getID(tok2idx, token):
    if token not in tok2idx:  # not seen during training
        return tok2idx["[UNK]"]
    else:
        return tok2idx[token]
```

```python
getID(tok2idx, "The")
```

```python
getID(tok2idx, "boat")
```

```python
getID(tok2idx, "asdf")
```

A sentence is then a sequence of IDs.

```python
sent = ["The", "old", "man", "the", "boat"]
x = [getID(tok2idx, t) for t in sent]
x
```

This is a natural representation, but it needs some extra handling because this vector can have variable length.

## Bag of words

To represent sentences with a fixed-length vector, one option is a *bag of words*: a vector where the index is the index of the token in our vocabulary, and the value at the index is the number of times it appears in the sentence.

```python
def bow(tok2idx, tokens):
    vec = [0 for _ in range(len(tok2idx))]
    for t in tokens:
        id = getID(tok2idx, t)
        vec[id] += 1
    return vec
```

```python
bow(tok2idx, tokens)
```
